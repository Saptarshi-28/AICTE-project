```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv("C:/Users/pandi/Desktop/project-hai/datasets/heart_disease_data.csv")
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X = data.drop(columns='target', axis=1)
Y = data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
param_grid={
    'n_estimators': [50, 100, 150, 200],  # Small dataset â†’ fewer trees
    'max_features': ['sqrt', 'log2', 0.5, None],  # Different feature selection methods
    'bootstrap':[True,False],
    'oob_score':[True,False]
}
```

```python
rfc=RandomForestClassifier()
```

```python
model2=GridSearchCV(rfc,param_grid)
```

```python
model2.fit(X_train,Y_train)
```

```python
#the above error is occuring because we try to use OOB without enabling bootstrap in some of the parameter grid combinations
```

```python
model2.best_params_
```

```python
X_test_prediction = model2.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
input_data = (57,0,0,120,354,0,1,163,1,0.6,2,0,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)
print(input_data_as_numpy_array.dtype)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
print(input_data_reshaped.dtype)

prediction = model2.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
